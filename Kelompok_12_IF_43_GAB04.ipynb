###**1. Umur Pelanggan dengan Kategori kartu**
Eksplorasi data yang kami lakukan menggunakan boxplot ini bertujuan untuk mengutahui keterkaitan jenis kartu yang dipilih dengan rentang usia pelanggan.

Rata-rata pemilihan jenis kartu dengan rentang usia terbanyak jatuh kepada jenis kartu blue, karena kartu jenis blue merupakan kartu yang paling tinggi tingkatannya diantara jenis kartu yang lain.

Terlihat jenis kartu blue, gold dan silver memiliki outliers setelah nilai maksimumnya, hal ini menunjukkan adanya usia pelanggan yang memilih kartu berjenis blue, gold dan silver yang bernilai terlalu jauh dari rentang usia pengguna lainnya. 

Berikut merupakan penjelasan distribusi-nya :

1. Boxplot Card Blue, distribusi data cenderung simetris yang berarti kemiringan dari distribusi data pada boxplot card blue bernilai 0
2. Boxplot Card Gold, distribusi data cenderung menjulur ke arah bawah yang berarti kemiringan dari distribusi data pada boxplot card gold bernilai negatif
3. Boxplot Card Platinum, distribusi data cenderung ke bawah yang berarti kemiringan dari distribusi pada boxplot card platinum bernilai negatif
4. Boxplot Card Silver, distribusi data cenderung menjulur ke arah atas yang berarti kemiringan dari distribusi data pada boxplot card silver bernilai positif










```
ggplot(data, aes(x = Card_Category , y = Customer_Age, color = Card_Category))+ 
geom_boxplot()+
scale_color_manual(values = c("#0000FF", "#D4AF37", "#FF0000", "#808080"))+ theme_classic()+
labs(x = "Card Category", y = "Customer Age")
```

##**HEATMAP**

```
ggplot(data, aes(Card_Category, Income_Category, fill = Customer_Age))+
geom_tile()+
labs(x = "Card Category", y = "Income Category", title="Distribution of Card Category & Income Category with Customer Age")+
scale_fill_distiller(palette = "Reds")+
theme_bw()+
theme(plot.title = element_text(size=12) ,axis.text.x= element_text(size=12), axis.text.y= element_text(size=12), axis.title=element_text(size=15))
```

Dari Heatmap diatas dapat kita ambil kesimpulan bahwa warna merah paling tua terdapat pada kisaran pemasukan sebesar $40K - $60K, Dimana pemasukan tersebut paling banyak ditemui di kalangan customer yang berumur 30-40 tahun.
